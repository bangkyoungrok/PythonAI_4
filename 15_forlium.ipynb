```python
# folium은 Leaflet.js를 이용해서 위치 좌표계 값을 시각화 할 수 있다
# !pip install folium
```

```python
# 보고자 하는 위치의 좌표계를 얻어내는 가장 단순한 방법은 구글맵을 이용하는 방법이다.
import folium
```

```python
# folium 라이브러리의 Map() 메소드의 location속성으로 지정한 위치를 중심으로 하는 zoom_start 옵션으로 지정할 배율을
# 사용하는 지도를 만든다.
#folium.Map(location=[위도, 경도],zoom_start=배율,[ tiles='지도 스타일'])
map_folium = folium.Map(location=[37.5662952,126.9757564],zoom_start=17)

# folium 라이브러리의 Marker() 메소드의 location 속성으로 지정한 위치에 마커를 만든다.
# popup 속성으로 Marker를 클릭했을 때 표시할 문자열을 지정할 수 있다.
# 작성된 Marker는 add_to() 메소드를 사용해서 지도에 추가한다.
folium.Marker(location=[37.56668469877371, 126.97839767630022],popup='서울특별시청').add_to(map_folium) 

folium.Marker(location=[37.5658362003022, 126.97514303974722],popup='덕수궁').add_to(map_folium) 
                                                                                

# folium의 save() 메소드로 작성된 지도 차트를 html 파일로 저장할 수 있다.
map_folium.save('./output/map.html')
map_folium
```

```python
map_folium = folium.Map(location=[37.5662952,126.9757564],zoom_start=17)

# Marker를 클릭했을때 팝업되는 말풍선의 크기가 마음에 들지 않으면 Popup() 메소드의 max_width 속성으로 변경하면 된다.
popup = folium.Popup("서어어어어우우울특벼어어얼시이처어엉", max_width=500).add_to(map_folium) 
folium.Marker(location=[37.56668469877371, 126.97839767630022],popup=popup).add_to(map_folium) 
folium.Marker(location=[37.5658362003022, 126.97514303974722],popup='덕수궁').add_to(map_folium) 

map_folium.save('./output/map.html')
map_folium
```

```python
map_folium = folium.Map(location=[37.5662952,126.9757564],zoom_start=17)

# icon 속성에 Icon 메소드로 마커의 색상을 지정할 수 있다.
folium.Marker(location=[37.56668469877371, 126.97839767630022],icon=folium.Icon(color='green')).add_to(map_folium) 
folium.Marker(location=[37.5658362003022, 126.97514303974722],popup='덕수궁').add_to(map_folium) 

map_folium.save('./output/map.html')
map_folium
```

```python
map_folium = folium.Map(location=[37.5662952,126.9757564],zoom_start=17)

# icon 속성에 Icon 메소드로 아이콘 모양을 지정할 수 있다.
# https://getbootstrap.com/docs/3.3/components/#glyphicons-glyph <- 아이콘 이미지를 볼 수 있다.
folium.Marker(location=[37.56668469877371, 126.97839767630022],
              icon=folium.Icon(color='green',icon ='save')).add_to(map_folium) 
folium.Marker(location=[37.5658362003022, 126.97514303974722],
              icon=folium.Icon(color='red',icon ='star')).add_to(map_folium) 

map_folium.save('./output/map.html')
map_folium
```

```python
map_folium = folium.Map(location=[37.5662952,126.9757564],zoom_start=17)

# 마커를 원모양으로 표시 이때 색은 html 색상코드 적어도 된다.
folium.CircleMarker(location=[37.56668469877371, 126.97839767630022],popup='서울특별시청',
                   radius=100, color='red').add_to(map_folium)
folium.CircleMarker(location=[37.5658362003022, 126.97514303974722],popup='덕수궁',
              radius=100,color='#66ccff',fill_color = '#ffff99').add_to(map_folium) 

map_folium.save('./output/map.html')
map_folium
```
