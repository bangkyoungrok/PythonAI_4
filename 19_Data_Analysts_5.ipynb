```python
# 집행일시에 NaN(결측치) 데이터가 있다. 해당 데이터를 확인해 본다.
df[df['집행일시'].isnull()]
```

```python
# 집행일시의 결측치 데이터를 얻어온다.
missing_df_row = df[df['집행일시'].isnull()].index[0]
```

```python
#  해당 행을 누락시키고 다시 df 데이터프레임에 저장한다.
print(df.shape)
df = df.drop(missing_df_row)
print(df.shape)
```

```python
# 집행일시 열을 이용해서 '연','월', '연원', '일', '시', '분', '요일' 파생 변수를 만든다.
df['집행일시']=pd.to_datetime(df['집행일시'])
df['연'] = df['집행일시'].dt.year.astype(int)
df['월'] = df['집행일시'].dt.month.astype(int)
df['연월'] = df['연'].astype(str)+'-'+df['월'].astype(str)
df['일'] = df['집행일시'].dt.day.astype(int)
df['시'] = df['집행일시'].dt.hour.astype(int)
df['분'] = df['집행일시'].dt.minute.astype(int)
df['요일'] = df['집행일시'].dt.dayofweek.astype(int) # 0 -> 월, 1 -> 화, 2 -> 수, 3 -> 목, 4 -> 금, 5 -> 토, 6 -> 일
df
```

### 집행목적 워드 클라우드

```python
from wordcloud import WordCloud
```

```python
df.head()
```

```python
# 집행목적이 누락값인 데이터를 확인한다.
df[df['집행목적'].isnull()]
```

```python
# 집행목적이 누락값인 데이터가 존재하므로 누락값을 공백으로 채워준다.
df['집행목적'] = df['집행목적'].fillna('')
df[df['집행목적'].isnull()]
```

```python
# 파이썬 문자열 함수 중에 특정 문자열을 문자열 사이에 삽인하는 join() 메소드를 사용해서 집행목적을 연결해서 워드클라우드를 
# 실행한다.
spend_gubun = ' '.join(df['집행목적'])
spend_gubun[:100]
```

```python
wordcloud = WordCloud(
    font_path = 'C:\\Users\\bkr03\\AppData\\Local\\Microsoft\\Windows\\Fonts\\NanumGothicCoding.ttf',
    background_color='white',
    relative_scaling=0.5
).generate(spend_gubun)
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

#### 집행장소 누락값

```python
# 집행장소가 누락값인 데이터를 확인한다.
df[df['집행장소'].isnull()]
```

```python
# 집행장소가 누락값인 데이터가 존재하므로 누락값을 공백으로 채워준다.
df['집행장소'] = df['집행장소'].fillna('')
df[df['집행장소'].isnull()]
```

```python
df.head()
```

```python
# 집행장소에 주소가 많이 보이기 떄문에 주소는 전처리를 통해 제거하고 상호명으라는 열을 만들어 집행장소에서 상호면을 가져온다.
# 상호명이 여는 괄호 앞쪽에 있는 것들은 여는 괄호 앞에 있는 것을 상호명으로 한다.
df['상호명'] = df['집행장소'].apply(lambda x: x.split('(')[0].strip())
df.head()
```

```python
# 상호명이 공백인 데이터의 집행목적을 확인한다.
df.loc[df['상호명']=='','집행목적'].value_counts()
```

```python
# 상호명이 '-'인 데이터의 집행목적을 확인한다.
df.loc[df['상호명']=='-','집행목적'].value_counts()
```

```python
# 상호명이 '대상자 소속부서'인 데이터의 집행목적을 확인한다.
df.loc[df['상호명']=='대상자 소속부서','집행목적'].value_counts()
```

```python
# 상호명이 공백, '-', '대상자 소속부서'인 데이터(잘못된 데이터)를 제거한다.
df_shop = df[(df['상호명'] != '')&(df['상호명'] != '-')&(df['상호명'] != '대상자 소속부서')]
df_shop
```
